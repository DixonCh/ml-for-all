```python
# Load data
import numpy as np
import matplotlib.pyplot as plt
data = np.linspace(-10.0, 10.0, num=100, endpoint=True);
```

```python
%matplotlib inline
plt.figure(figsize=(15,5))
plt.plot(data)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid(True)
plt.title("Raw data")
plt.show()
```

```python
# Linear (Identity) Function

def linear(x, Derivative=False):
    if not Derivative:
        return x
    else:
        return np.ones_like(x)
    
output = linear(data);
output1 = linear(data, True);

plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.plot(output)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid(True)
plt.title("Linear Function")


plt.subplot(1,2,2)
plt.plot(output1 )
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid(True)
plt.title("Linear Function (Derivative)")

plot.show()
```

```python
#The Sigmoid (or Fermi) Function
def sigmoid(x,Derivative=False):
    if not Derivative:
        return 1 / (1 + np.exp (-x))
    else:
        out = sigmoid(x)
        return out * (1 - out)
    
output = sigmoid(data);
output1 = sigmoid(data, True);

plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.plot(output)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid(True)
plt.title("Sigmoid Function")


plt.subplot(1,2,2)
plt.plot(output1)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid(True)
plt.title("Sigmoid Function (Derivative)")

plot.show()
```

```python
# Hyperbolic Tangent Function
import math

def htan(x,Derivative=False):
    if not Derivative:
#         return np.array([math.tanh(v) for v in x])
        return np.tanh(x)
    else:
        return (1 - np.power(np.tanh(x),2))

                            
output = htan(data);
output1 = htan(data, True);

plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.plot(output)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid(True)
plt.title("Hyperbolic Tangent Function")


plt.subplot(1,2,2)
plt.plot(output1)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid(True)
plt.title("Hyperbolic Tangent Function (Derivative)")

plot.show()
```

```python
#Gaussian Function
def gaussian(x, Derivative=False):
    if not Derivative:
        return np.exp(-x**2)
    else:
        return -2 * x * np.exp(-x**2)

                            
output = gaussian(data);
output1 = gaussian(data, True);

plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.plot(output)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid(True)
plt.title("Gaussian Function")


plt.subplot(1,2,2)
plt.plot(output1)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid(True)
plt.title("Gaussian Function (Derivative)")

plot.show()
```

```python
# Step (or Heaviside) Function
def step(x, Derivative=False):
    if not Derivative:
        temp = []
        for v in x:
            if v > 0:
                temp.append(1)
            else:
                temp.append(0)
        return np.array(temp)
    else:
        return np.zeros_like(x)
                            
output = step(data);
output1 = step(data, True);

plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.plot(output)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid(True)
plt.title("Step Function")


plt.subplot(1,2,2)
plt.plot(output1, '*')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid(True)
plt.title("Step Function (Derivative)")

plot.show()
```

```python
# Ramp Function

def ramp(x, Derivative=False, T1=0, T2=0):
    if T2 is 0:
        T2 = np.max(x)
    out = np.ones(x.shape)
    ids = ((x < T1) | (x > T2))
    if not Derivative:
        out = ((x - T1)/(T2-T1))
        out[(x < T1)] = 0
        out[(x > T2)] = 1
        return out
    else:
        out[ids]=0
        return out

output = ramp(data, T1 =-2, T2 =3);    
output1 = ramp(data, T1 =-2, T2 =3, Derivative = True);

plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.plot(output)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid(True)
plt.title("Ramp Function")


plt.subplot(1,2,2)
plt.plot(output1)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid(True)
plt.title("Ramp Function (Derivative)")

plot.show()
```

```python
# Rectified Linear Unit (ReLU)
def relu(x, Derivative=False):
    if not Derivative:
        return np.maximum(0,x)
    else:
        out = np.ones(x.shape)
        out[(x < 0)]=0
        return out

    
output = relu(data);    
output1 = relu(data,  Derivative = True);

plt.figure(figsize=(15,5))

ax1 = plt.subplot(1,2,1)
plt.axis([-10, 10, -10,10])
plt.plot(output)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid(True)
plt.title("ReLU Function")


ax2 = plt.subplot(1,2,2, sharex=ax1)
plt.plot(output1)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid(True)
plt.title("ReLU Function (Derivative)")

plot.show()
```
